# üè° House Prices: Feature Engineering & Analysis
This notebook performs feature engineering on the Ames Housing dataset to prepare for predictive modeling.

## üì¶ Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
pd.set_option('display.max_columns', None)
sns.set(style='whitegrid')
```

## üì• Load Dataset

```python
# Load the dataset
df = pd.read_csv(r"C:\Users\line\OneDrive\Documents\Real Estate Data.csv")
df.head()
```

## üîç Missing Values Analysis

```python
# Check missing values
missing = df.isnull().sum()
missing[missing > 0].sort_values(ascending=False)
```

## üõ† Handling Missing Data

```python
# Fill NA values for categorical features with 'None' and numerical with 0
na_fill_none = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 
                'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'BsmtQual', 
                'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'MasVnrType']
na_fill_zero = ['GarageYrBlt', 'GarageArea', 'GarageCars', 'MasVnrArea', 
                'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath']

for col in na_fill_none:
    df[col] = df[col].fillna('None')
for col in na_fill_zero:
    df[col] = df[col].fillna(0)

# Fill remaining with mode
df.fillna(df.mode().iloc[0], inplace=True)
```

## üß† Feature Engineering

```python
# New features
df['TotalSF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']
df['TotalBath'] = df['FullBath'] + (0.5 * df['HalfBath']) + df['BsmtFullBath'] + (0.5 * df['BsmtHalfBath'])
df['Age'] = df['YrSold'] - df['YearBuilt']
df['RemodAge'] = df['YrSold'] - df['YearRemodAdd']
df['HasPool'] = df['PoolArea'].apply(lambda x: 1 if x > 0 else 0)
df['HasGarage'] = df['GarageArea'].apply(lambda x: 1 if x > 0 else 0)
df[['TotalSF', 'TotalBath', 'Age', 'RemodAge', 'HasPool', 'HasGarage']].head()
```

## üìä Feature Correlation with SalePrice

```python
corr = df.corr(numeric_only=True)
top_corr = corr['SalePrice'].abs().sort_values(ascending=False).head(10)
sns.heatmap(df[top_corr.index].corr(), annot=True, cmap='coolwarm')
plt.title('Top Correlated Features with SalePrice')
plt.show()
```
